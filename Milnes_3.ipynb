```python
import pandas as pd
import seaborn as sns
```

```python
av = pd.read_csv('avocado.csv')
av.head()
```

```python
av.info()
```

```python
av['Date'] = pd.to_datetime(av['Date'],format='%Y-%m-%d')
av.info()
```

```python
av_bags = av.query('region == "Albany"').query('year == 2015')[['Date','Total Bags','Small Bags']]
av_bags.head()
```

```python
av_bags_melt = pd.melt(av_bags, id_vars='Date', value_vars=['Total Bags','Small Bags'], var_name='Bags', value_name='Count')
av_bags_melt
```

```python
sns.relplot(data=av_bags_melt, kind='line', x='Date', y='Count', hue='Bags')
```

```python
av['Sales Volume'] = pd.qcut(av['Total Volume'], q=4, labels=['Poor','Modest','Good','Excellent'])
av.head()
```

```python
sns.countplot(data=av, x='year', hue='Sales Volume')
```

```python
av_bags_melt2 = pd.melt(av, id_vars='year', value_vars=['Small Bags','Large Bags','XLarge Bags'], var_name='Bags', value_name='Count')
av_bags_melt2
```

```python
sns.relplot(data=av_bags_melt2, kind='line', x='year', y='Count', hue='Bags')
```

## It seems that the small bags sell the best.
